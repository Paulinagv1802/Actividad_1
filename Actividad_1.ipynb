```python
#Ejercicio 4

# Esta clase se encarga de realizar los cálculos matemáticos del ejercicio
class Calcular_Edades:

    @staticmethod
    def calcular_edalber(edjuan):
        return (2 * edjuan) / 3  #Formula para calcular la edad de Alberto

    @staticmethod
    def calcular_edana(edjuan):
        return (4 * edjuan) / 3 #Formula para calcular la edad de Ana

    @staticmethod
    def calcular_edmama(edjuan, edalber, edana):
        return edjuan + edalber + edana

edjuan = int(input("Ingrese la edad de juan: "))
edalber = Calcular_Edades.calcular_edalber(edjuan)
edana = Calcular_Edades.calcular_edana(edjuan)
edmama = Calcular_Edades.calcular_edmama(edjuan, edalber, edana)

# Resultado
print(f"Las edades son: Alberto = {edalber}, Juan = {edjuan}, Ana = {edana}, Mamá = {edmama}")
```

```python
#Ejercicio 5

#Primero importamos la libreria

class Calculos:
   def __init__(self, x_inicial, y_inicial):
    self.suma = 0
    self.x = x_inicial
    self.y = y_inicial

   def ejecutar(self):
      self.suma += self.x      # Suma = suma + x
      self.x += self.y ** 2    # x = x + y**2
      self.suma += self.x / self.y     #suma += x/y
      return self.suma

#Objeto

proceso = Calculos(20, 40)  #valores que toma x y y
print(f"El valor inicial de x es: {proceso.x}")
print(f"El valor inicial de y es: {proceso.y}")
print(f"El valor de la suma es: {proceso.ejecutar()}")
```

```python
#Ejercicio 12

class Nomina:

  @staticmethod
  def calculo_salario_bruto(horas_trabajadas, valor_hora):
    return horas_trabajadas * valor_hora


  @staticmethod
  def calculo_retencion(salario_bruto, porcentaje_retencion):
    return salario_bruto * porcentaje_retencion


  @staticmethod
  def calculo_salario_neto(salario_bruto, valor_retencion):
    return salario_bruto - valor_retencion


#Definimos las variables
horas_trabajadas = 48
valor_hora = 5000
retencion = 12.5
porcentaje_retencion = retencion / 100



#calculos usando la clase nomina
salario_bruto = Nomina.calculo_salario_bruto(horas_trabajadas, valor_hora)
valor_retencion = Nomina.calculo_retencion(salario_bruto, porcentaje_retencion)
salario_neto = Nomina.calculo_salario_neto(salario_bruto, valor_retencion)



# Impresión de resultados
print(f"Salario Bruto: ${salario_bruto:,.0f}")
print(f"Retención en la Fuente: ${valor_retencion:,.0f}")
print(f"Salario Neto: ${salario_neto:,.0f}")
```

```python
#Ejercicio 14

class CalculadoraPotencias:

    @staticmethod
    def calcular_cuadrado(numero):
        return numero ** 2

    @staticmethod
    def calcular_cubo(numero):
        return numero ** 3


numero = float(input("Ingrese un número: "))


# Llamada a los métodos estáticos de la clase Calculos
cuadrado = CalculadoraPotencias.calcular_cuadrado(numero)
cubo = CalculadoraPotencias.calcular_cubo(numero)


print(f"El cuadrado es: {cuadrado}")
print(f"El cubo es: {cubo}")
```

```python
#Ejercicio 17

#Importamos la libreria math
import math

class CalculoCirculo:

  @staticmethod
  def calculo_area(radio):    #area = pi * radio**2
    return math.pi * radio ** 2

  @staticmethod
  def calculo_longitud_circurferencia(radio):    #Longitud = 2*pi*radio
    return 2 * math.pi * radio

radio = float(input("Ingrese el radio del círculo: "))

#Llamamos a la clase

area = CalculoCirculo.calculo_area(radio)
longitud = CalculoCirculo.calculo_longitud_circurferencia(radio)


print(f"El área es: {area:.2f}")
print(f"La longitud es: {longitud:.2f}")

```
